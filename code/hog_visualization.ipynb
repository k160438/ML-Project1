```python
import os
import sys
from skimage.feature import hog
import numpy as np
import cv2
import matplotlib.pyplot as plt
% matplotlib inline

X_train = np.load('data/X_train.npy')
X_test = np.load('data/X_test.npy')


```

```python
positive = cv2.imread('data/visual/report_use/positive_2002_08_18_big_img_181.jpg')
negative = cv2.imread('data/visual/report_use/negative_2002_08_18_big_img_181.jpg')
positive = cv2.cvtColor(positive, cv2.COLOR_BGR2RGB)
negative = cv2.cvtColor(negative, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
plt.imshow(positive)
plt.title('Bounding boxes visualization of \nPositive Sample')
plt.subplot(1,2,2)
plt.imshow(negative)
plt.title('Bounding boxes visualization of \nNegative Sample')
plt.savefig('data/visual/report_use/bounding_boxes.jpg')
```

```python
img = X_train[6]
feature, hog_img = hog(img, 9, (16, 16), (2, 2), visualise=True, feature_vector=True)
hog_img = hog_img/np.max(hog_img)
print(img.shape, hog_img.shape)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,4))
plt.subplot(1,2,1)
plt.title('Raw Image')
plt.imshow(img)
plt.subplot(1,2,2)
plt.title('HOG Feature Visualization')
plt.imshow(hog_img,cmap='gray')
plt.savefig('data/visual/report_use/hog_visualization.jpg')
```
